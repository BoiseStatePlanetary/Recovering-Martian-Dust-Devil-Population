```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from statsmodels.robust import mad
```

From Greeley+ (2006 - https://agupubs.onlinelibrary.wiley.com/doi/10.1029/2006JE002743):

$\tau = \ln \left( \dfrac{I_{\rm ground} - I_{\rm sky}}{I_{\rm DD} - I_{\rm sky}} \right)$,
where $\tau$ is optical depth, $I_{\rm ground}/I_{\rm sky}/I_{\rm DD}$ is the pixel value near the ground/in the sky/in a dust devil.

Presumably, we could have seen a dust devil if it had a contrast at least as great as the typicaly dimmest pixel in the sky ($\langle I_{\rm sky} \rangle - \sigma_{I_{\rm sky}}$) as compared to the typical brightest pixel ($\langle I_{\rm sky} \rangle - \sigma_{I_{\rm sky}}$). 

```python
# The sky
sky_image_file = "../figures/C000M0065_602302474EJP_F0000_0461M2_just_sky_BW.JPG"

sky_img = Image.open(sky_image_file)
np_sky_img = np.array(sky_img)
  
plt.imshow(sky_img)
```

```python
# max value - exclude the white portion of the image
max_val = 225

hst, bin_edges = np.histogram(np_sky_img, bins=len(np.unique(np_sky_img)))
bins = 0.5*(bin_edges[1:] + bin_edges[0:-1])
plt.plot(bins[bins < max_val], hst[bins < max_val])

I_sky_mad = mad(np_img[np_sky_img < max_val])
I_sky_med = np.median(np_img[np_sky_img < max_val])
print(I_sky_med, I_sky_mad)
```

```python
# The ground
ground_image_file = "../figures/C000M0065_602302474EJP_F0000_0461M2_just_ground_BW.jpg"

ground_img = Image.open(ground_image_file)
np_ground_img = np.array(ground_img)
  
plt.imshow(ground_img)
```

```python
hst, bin_edges = np.histogram(np_ground_img, bins=len(np.unique(np_ground_img)))
bins = 0.5*(bin_edges[1:] + bin_edges[0:-1])
plt.plot(bins[bins < max_val], hst[bins < max_val])

I_ground_mad = mad(np_ground_img[np_ground_img < max_val])
I_ground_med = np.median(np_ground_img[np_ground_img < max_val])
print(I_ground_med, I_ground_mad)
```

```python
num = (I_ground_med - I_sky_med)

# This value assumes a dust devil could NOT have been darker than 3-sigma_I_sky. 
# Otherwise, we would have spotted it.
denom = (-3.*I_sky_mad)

max_tau = np.log(num/denom)
print(max_tau)
```
