```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit

aspect_ratio = 16./9
BoiseState_blue = '#0033A0'
BoiseState_orange = '#D64309'
```

```python
def Lorentz_profile(t, t0, baseline, Pact, diameter, velocity, impact_parameter):
    r = np.sqrt(impact_parameter**2 + (velocity*(t - t0))**2)
    
    return baseline - Pact/(1. + (2.*r/diameter)**2)

t = np.linspace(-10., 10., 100)
baseline = 100.
Pact = 10.
diameter = 10.
velocity = 10.
impact_parameter = 10.

plt.plot(t, Lorentz_profile(t, 0., baseline, Pact, diameter, velocity, impact_parameter))
```

```python
pressure_profile_filename = "41561_2020_534_Fig5_HTML_c.txt"
wind_profile_filename = "41561_2020_534_Fig5_HTML_d.txt"

pressure_data = np.genfromtxt(pressure_profile_filename, delimiter=", ", names=["time", "pressure"])
wind_data = np.genfromtxt(wind_profile_filename, delimiter=", ", names=["time", "windspeed"])

# Uniquify the data
_, ind = np.unique(pressure_data["time"], return_index=True)
pressure_times, pressures = pressure_data["time"][ind], pressure_data["pressure"][ind]

_, ind = np.unique(wind_data["time"], return_index=True)
wind_times, windspeeds = wind_data["time"][ind], wind_data["windspeed"][ind]

mn = np.min(pressure_times)
pressure_x = (pressure_times - mn)*3600.
wind_x = (wind_times - mn)*3600.

fig = plt.figure(figsize=(6*aspect_ratio, 6))
ax1 = fig.add_subplot(111)
ax2 = ax1.twinx()

ax1.plot(wind_x, windspeeds, color=BoiseState_blue, lw=6)
ax1.tick_params(labelsize=24)
ax1.tick_params(axis='y', labelcolor=BoiseState_blue)
ax1.set_xlabel("$t\,\\left({\\rm sec}\\right)$", fontsize=36)
ax1.set_ylabel("$V\,\\left({\\rm m\ s^{-1}}\\right)$", fontsize=36, color=BoiseState_blue)

ax2.plot(pressure_x, pressures, color=BoiseState_orange, lw=6)
ax2.tick_params(labelsize=24)
ax2.tick_params(axis='y', labelcolor=BoiseState_orange)
ax2.set_ylabel("$P\,\\left({\\rm Pa}\\right)$", fontsize=36, color=BoiseState_orange)

baseline = np.mean(pressures[(pressure_x < 20.) | (pressure_x > 80.)])
pressure_err = np.std(pressures[(pressure_x < 20.) | (pressure_x > 80.)])

transverse_velocity = np.mean(windspeeds[wind_x > 80.])
velocity_err = np.std(windspeeds[wind_x > 80.])
print(transverse_velocity, velocity_err)

ax1.axvline(45)
ax1.axvline(57)
```

```python
# baseline, Pact, diameter, velocity, impact_parameter
p0 = [pressure_x[np.argmin(pressures)], 702.3, 10., 30., transverse_velocity, 0.]

ind = (pressure_x > 45) & (pressure_x < 57)

popt, pcov = curve_fit(Lorentz_profile, pressure_x[ind], pressures[ind], 
                       sigma=pressure_err*np.ones_like(pressures[ind]), p0=p0)
print(popt)

plt.plot(pressure_x, pressures)
plt.plot(pressure_x, Lorentz_profile(pressure_x, *popt))

t0 = popt[0]
```

```python
t = pressure_x - t0
pressure_profile = pressures - baseline
plt.plot(t, pressure_profile)
```
